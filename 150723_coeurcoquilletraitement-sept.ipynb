```python
import coeurcoquilledisp

breath=[]
#coeur germanium, coquille Si
for x in fraction:
    dispcoeurcoq=coeurcoquilledisp.dispersion(E_coq=169e9, nu_coq=0.23, rho_coq=2330, 
                                          a_coq=100.0*1e-9,E_coeur=132e9, nu_coeur=0.19, rho_coeur=5320, 
                                          a_coeur=x*100.0*1e-9,f_min=0.1, f_max=50e9, f_pas=1e7)
    trac=dispcoeurcoq.relation()
    breath.append(trac)
    
breathinv=[]
#coeur Si, coquille Ge
for x in fraction:
    dispcoeurcoqinv=coeurcoquilledisp.dispersion(E_coq=132e9, nu_coq=0.19, rho_coq=5320, 
                                          a_coq=100.0*1e-9,E_coeur=169e9, nu_coeur=0.23, rho_coeur=2330, 
                                          a_coeur=x*100.0*1e-9,f_min=0.1, f_max=50e9, f_pas=1e7)
    tracinv=dispcoeurcoqinv.relation()
    breathinv.append(tracinv)
```

```python
def stars(p):
   if p < 0.0001:
       return "****"
   elif (p < 0.001):
       return "***"
   elif (p < 0.01):
       return "**"
   elif (p < 0.05):
       return "*"
   else:
       return "-"
```
